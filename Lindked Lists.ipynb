```python
# List Node Class.
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def mergeSort(head):
    # Write your code here.
    if head==None or head.next==None :
        return head 

    mid=middle(head)
    left=head
    right=mid.next 
    mid.next=None 

    left=mergeSort(left)
    right=mergeSort(right)
    head=merge(left,right)
    return head

def middle(head):
    slow,fast=head,head.next
    while fast and fast.next:
        slow=slow.next 
        fast=fast.next.next
    return slow 

def merge(left,right):
    if left==None:
        return right
    if right==None:
        return left 

    if left.data<=right.data:
        head=left 
        left=left.next
    else:
        head=right
        right=right.next

    curr=head 
    while left and right :
        if left.data<=right.data:
            curr.next=left 
            left=left.next 
        else:
            curr.next=right
            right=right.next
        curr=curr.next 

    if left==None:
        curr.next=right 
    else:
        curr.next=left

    return head

def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

def printll(head):
    while head:
        print(head.data, end=' ')
        head = head.next
    print()

# Main.
# Read the link list elements including -1.
arr=[1,2,4,1,5,-1]
# Create a Linked list after removing -1 from list.
l = ll(arr[:-1])
l = mergeSort(l)
printll(l)
```

```python
#From Sorted List
def removeDuplicates(head: Node) -> Node:
    # Write your code here
    curr=head
    while curr.next:
        if curr.data==curr.next.data:
            curr.next=curr.next.next
        else:
            curr=curr.next
    return head
```

```python
#From Unsorted List
def removeDuplicates(head):
    # Write your code here
    d=set()
    curr=head
    prev=None  
    while curr:
        if curr.data not in d:
            d.add(curr.data)
            prev=curr
            curr=curr.next
        else:
            curr=curr.next 
            prev.next=curr
    return head
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def length(head):
    l = 0
    while head:
        head = head.next
        l += 1
    return l

def findIntersection(firstHead, secondHead):
    fl = length(firstHead)
    sl = length(secondHead)

    while fl > sl:
        firstHead = firstHead.next
        fl -= 1
    while sl > fl:
        secondHead = secondHead.next
        sl -= 1

    while firstHead is not None and firstHead != secondHead:
        firstHead = firstHead.next
        secondHead = secondHead.next

    if firstHead is None:
        return -1  # No intersection found

    return firstHead.data

# Example usage:
# Create two linked lists with a common intersection point
commonNode = Node(7)
firstHead = Node(1)
firstHead.next = Node(2)
firstHead.next.next = Node(3)
firstHead.next.next.next = commonNode

secondHead = Node(10)
secondHead.next = Node(9)
secondHead.next.next = commonNode

intersection_point = findIntersection(firstHead, secondHead)
print("Intersection point:", intersection_point)
```
